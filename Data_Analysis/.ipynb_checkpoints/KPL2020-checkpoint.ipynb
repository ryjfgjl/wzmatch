```python
import pandas as pd
#import tushare as ts
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
import pymysql
conn = pymysql.connect(host='localhost',port=3306,db='wzmatch',user='root',passwd='root',charset='utf8')
def sele(sql):
    cur = conn.cursor()
    cur.execute(sql)
    result = list(cur.fetchall())
    result = [list(r) for r in result]
    conn.commit()
    cur.close()
    
    return result
```

```python
sql = 'select * from kpl2020_autumn'
data = sele(sql)
```

```python
frame = pd.DataFrame(data, columns=['id', 'matchTime', 'teamA', 'teamB', 'score', 'scoreA', 'scoreB', 'BO'])
frame = frame[frame['BO']=='BO5']
frame[1:10]
```

```python
# clean
gy = frame['score'].groupby(frame['score'])

# score
def f1(x):
    return x.split('-')[0].strip()
def f2(x):
    return x.split('-')[-1].strip()
frame['scoreA'] = frame['score'].apply(f1)
frame['scoreB'] = frame['score'].apply(f2)
# BO
frame['BO'] = 5
frame
```

```python
sql = 'select * from kpl2020_spring'
spring_data = sele(sql)
spring_frame = pd.DataFrame(spring_data, columns=['id', 'matchTime', 'teamA', 'teamB', 'score', 'scoreA', 'scoreB', 'BO'])

# score
def f1(x):
    return int(x.split('-')[0].strip())
def f2(x):
    return int(x.split('-')[-1].strip())
spring_frame['scoreA'] = spring_frame['score'].apply(f1)
spring_frame['scoreB'] = spring_frame['score'].apply(f2)
# BO
spring_frame.loc[spring_frame[spring_frame['BO']=='BO5'].index,'BO'] = 5
spring_frame.loc[spring_frame[spring_frame['BO']=='BO7'].index,'BO'] = 7
spring_frame
```

```python
f1 = spring_frame[['teamA','scoreA']]
f2 = spring_frame[['teamB','scoreB']]
scores = f1 + f2
scores
```
