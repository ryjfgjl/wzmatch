```python
import pymysql
conn = pymysql.connect(host='localshot',port=3306,db='wzmatch',user='root',passwd='root',charset='utf8')
def sele(sql):
    cur = conn.cursor()
    cur.execute(sql)
    result = list(cur.fetchall())
    result = [list(r) for r in result]
    conn.commit()
    cur.close()
    
    return result
```

```python
sql = 'select * from kpl2020_autumn'
data = sele(sql)
data
```

```python
frame = pd.DataFrame(data, columns=['id', 'matchTime', 'teamA', 'temeB', 'score', 'scoreA', 'scoreB', 'BO'])
frame = frame[frame['BO']=='BO5']
frame[1:10]
```
